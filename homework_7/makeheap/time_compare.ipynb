```python
from main import makeheap_n_log_n, makeheap
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(42)
sizes = list(range(0, 10_000, 100))
n_for_step = 20
n_log_n_time_mean = []
n_time_mean = []

for size in sizes:
    n_log_n_time = []
    n_time = []
    for _ in range(n_for_step):
        print(f'\r{size}', end='')
        arr1 = rng.random([size, 2]).tolist()
        arr2 = arr1.copy()
        n_log_n_time.append(makeheap_n_log_n(arr1)[1])
        n_time.append(makeheap(arr2)[1])
    n_log_n_time_mean.append(sum(n_log_n_time)/n_for_step)
    n_time_mean.append(sum(n_time)/n_for_step)
```

```python
plt.plot(sizes, n_log_n_time_mean, label='n log n')
plt.plot(sizes, n_time_mean, label='n')
plt.xlabel('array size')
plt.ylabel('time')
plt.legend()
plt.show()
```
