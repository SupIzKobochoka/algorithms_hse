Что делаем? Берём array, кладём его в array_heap, дальше справо налево проходимся по всем элементам и делаем просеивание вниз этих элементов (если есть потомки, но это можно опустить)  
Получается для уровня глубины дерева h сложность составляет
$$
(h_{max}-h_{current})\ast2^{h_{current}-1}
$$  
Где первый множитель - это число операций при просеивании вниз, второй - число элементов на данном уровне

Значит для всех уровней сложность составляет
$$
\sum_{i=1}^{h}{(h-i)*2^{i-1}}, \text{где }h=\log_2{(n+1)}\\
     \frac{h}{2}\sum_{i=1}^{h}{2^{i}}-\frac{1}{2}\sum_{i=1}^{h}{i*2^{i}}\\
     \text{(по сумме геометрической прогрессии и другой, которую я честно в интернете нашёл)}\\
     h*(2^h-1)-(h-1)2^{h}-1\\
     h*2^h-h-h*2^h+2^{h}-1\\
     2^{h}-h+1=2^{log_2{(n+1)}}-log_2{(n+1)}-1\\
     (n+1)-log_2{(n+1)}-1\text{ - ну и тут очевидно, что сложность итоговая O(n)}\\
     
$$
